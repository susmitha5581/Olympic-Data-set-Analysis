```python
df_Agegroups = df
df_Agegroups["Age"] = df_Agegroups["Age"].astype(int)

def getAgeGroup(Age):
    if Age > 0 and Age < 20:
        return 1
    elif Age >= 20 and Age < 30:
        return 2
    elif Age >= 30 and Age < 40:
        return 3
    elif Age >= 40 and Age < 50:
        return 4
    elif Age >= 50 and Age < 60:
        return 5
    elif Age >= 60:
        return 6
    
df_Agegroups["Age_group"] =  df_Agegroups["Age"].apply(lambda x: getAgeGroup(x) )

plt.figure(figsize=(20, 10))
plt.title('count of respondents in each age group')
sns.countplot(df_Agegroups["Age_group"])
plt.show()
```

# Question 2: Which sports recived highest Medals in People's Republic of China with the Pie chart representation

```python
China_Medal = df[(df.Year == max_year) & (df.Medal==3) & (df.Team == "People's Republic of China")]
China_Medal = China_Medal[['Sport','Medal']].groupby('Sport').count()
China_Medal.reset_index(inplace=True)
Top_sports = China_Medal.sort_values('Medal', ascending=False)
Top_sports.head(10)
```

```python
fig1, ax1 = plt.subplots()

colors = ['Red', 'Yellow', 'Pink', 'Green', 'Orange']

#explsion
explode = (0.05,0.05,0.05,0.05,0.05)

ax1.pie(Top_sports.head()['Medal'], colors = colors, labels= Top_sports.head()['Sport'], autopct='%1.1f%%', startangle=90
       , pctdistance=0.85, explode = explode)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.axis('equal')  
plt.tight_layout()
plt.show()
```

# Question 3: Top 10 Players Who won highest Medals in the olympics with Pie Chart Representation

```python
df_medal_holders = df[df.Medal !=0]
df_medal_holders['Count_Of_Medals'] = 1
df_medal_holders.head()
```

```python
df_highest_medals = df_medal_holders[['Name', 'Year', 'Team', 'Count_Of_Medals']].groupby(['Name','Year','Team']).sum().sort_values('Count_Of_Medals', ascending=False)
df_highest_medals = df_highest_medals.groupby(['Name','Team']).sum().sort_values('Count_Of_Medals', ascending=False).head(10)
df_highest_medals.reset_index(inplace=True)
df_highest_medals
```

```python
from matplotlib import pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
Team = df_highest_medals.Name
Count_of_Medal = df_highest_medals.Count_Of_Medals
ax.pie(Count_of_Medal, labels = Team,autopct='%1.2f%%')
plt.show()
```
